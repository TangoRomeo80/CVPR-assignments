```python
# ignore information messgaes from tensorflow and check gpu


import os
import tensorflow as tf
os.environ['TFF_CPP_MIN_LOG_LEVEL'] = '2'

gpu_device = tf.config.experimental.list_physical_devices('GPU')
print(f"Number of GPU = {len(gpu_device)}")
tf.config.experimental.set_memory_growth(gpu_device[0], True)
```

```python
#import dependencies

from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import *
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam

import numpy as np
import matplotlib.pyplot as plt
```

```python
#load datasets

(X_train, _),(_, _) = mnist.load_data()
```

```python
X_train.shape
```

```python
#normalize data betweeen (-1, 1)

X_train = (X_train-127.5)/127.5
print(X_train.min())
print(X_train.max())
```

```python
#Batch and parameters declarations

EPOCHS = 500
BATCH_SIZE = 256
HALF_BATCH = 128
NO_OF_BATCHES = int(X_train.shape[0]/BATCH_SIZE)

NOISE_DIM = 100

adam = Adam(learning_rate = 2e-4, beta_1 = 0.5)
```

```python
#Model generator by upsampling

generator = Sequential() #sequential model

generator.add(Dense(units = 7 * 7 * 128, input_shape = (NOISE_DIM,)))#Dense layer definition
generator.add(Reshape((7,7,128)))
generator.add(LeakyReLU(0.2))
generator.add(BatchNormalization())

#increasing sample size((7,7,128) to (14,14,64))

generator.add(Conv2DTranspose(64, (3, 3), strides = (2, 2), padding = 'same'))
generator.add(LeakyReLU(0.2))
generator.add(BatchNormalization())

#increasing sample size((14,14,64) to (28,28,1))

generator.add(Conv2DTranspose(1, (3, 3), strides = (2, 2), padding = 'same', activation = 'tanh'))

#compile the generator
generator.compile(loss = keras.losses.binary_crossentropy, optimizer = adam)

#print summary of generator
generator.summary()
```

```python
#Model generator by downsampling

discriminator = Sequential()

#decreasing sample size((28,28,1) to (14,14,64))

discriminator.add(Conv2D(64, kernel_size = (3, 3), strides = (2, 2), padding = 'same', input_shape = (28, 28, 1)))
discriminator.add(LeakyReLU(0.2))

#decreasing sample size((14,14,64) to (7,7,128))

discriminator.add(Conv2D(128, kernel_size = (3, 3), strides = (2, 2), padding = 'same'))
discriminator.add(LeakyReLU(0.2))

#converting sample to scalar((7,7,128) to 6272)

discriminator.add(Flatten())
discriminator.add(Dense(100))
discriminator.add(LeakyReLU(0.2))

#output layer with one output

discriminator.add(Dense(1, activation = 'sigmoid'))

#compile the discriminator

discriminator.compile(loss = keras.losses.binary_crossentropy, optimizer = adam)

#print summary of discriminator

discriminator.summary()
```

```python
#Combining the generator and discriminator

discriminator.trainable = False

#set input dimensions

g_input = Input(shape = (NOISE_DIM, ))

#generate image from generator

gen_img = generator(g_input)

#check with discriminator

g_output = discriminator(gen_img)

#using Functional API to create final model

model = Model(g_input, g_output)

#compile model

model.compile(loss = keras.losses.binary_crossentropy, optimizer = adam)

#summurize model

model.summary()
```

```python
#reshape train data

X_train = X_train.reshape(-1, 28, 28, 1)
X_train.shape
```

```python
#display or save images based on requirements
def display_images(epoch, samples = 25):
    noise = np.random.normal(0, 1, size = (samples, NOISE_DIM))
    
    generated_img = generator.predict(noise)
    plt.figure(figsize = (10, 10))
    for i in range(samples):
        plt.subplot(5, 5, i + 1)
        plt.imshow(generated_img[i].reshape(28, 28), cmap = 'binary')
        plt.axis('off')
    
    script_dir = os.getcwd()
    out_dir = os.path.join(script_dir, 'Output/')
    plt.savefig(out_dir + '{}.png'.format(epoch))
    plt.show()
```

```python
#Training loop

d_losses = []
g_losses = []

#Running training loops for epochs

for epoch in range(EPOCHS):
    epoch_d_loss = 0.0
    epoch_g_loss = 0.0
    
    #batch gradient descent
    
    for step in range(NO_OF_BATCHES):
        
        #Step 1: train discriminator
        discriminator.trainable = True
        
        #get real data
        idx = np.random.randint(0, 60000, HALF_BATCH)
        real_imgs = X_train[idx]
        
        #get fake data
        noise = np.random.normal(0, 1, size = (HALF_BATCH, NOISE_DIM))
        
        fake_imgs = generator.predict(noise)
        
        #labels
        real_y = np.ones((HALF_BATCH, 1))*0.9
        fake_y = np.zeros((HALF_BATCH, 1))
        
        #train discrimintaor
        d_loss_real = discriminator.train_on_batch(real_imgs, real_y)
        d_loss_fake = discriminator.train_on_batch(fake_imgs, fake_y)
        
        d_loss = (0.5 * d_loss_real) + (0.5 * d_loss_fake)
        
        epoch_d_loss += d_loss
        
        #Step 2: train generator
        discriminator.trainable = False
        
        noise = np.random.normal(0, 1, size = (BATCH_SIZE, NOISE_DIM))
        
        grnd_truth_y = np.ones((BATCH_SIZE, 1))
        
        g_loss = model.train_on_batch(noise, grnd_truth_y)
        
        epoch_g_loss += g_loss 
        
    #print results
        
    print(f'Epoch: {epoch + 1}, Discriminator loss: {epoch_d_loss/NO_OF_BATCHES}, Generator loss: {epoch_g_loss/NO_OF_BATCHES}')
    d_losses.append(epoch_d_loss/NO_OF_BATCHES)
    g_losses.append(epoch_g_loss/NO_OF_BATCHES)    
        
        
    if(epoch + 1) % 10 == 0.0:
#       generator.save("generator.h5")
        display_images(epoch + 1, 25)
```
